## Observable Trends

From this analysis, I was able to investigate the weather trends in relation to latitude. Before we look at the data, it is necessary to acknowledge about geographical characteristics in terms of Northern and Southern Hemisphere: there is more landmass in the Northern Hemisphere, whereas there is more ocean in the Southern Hemisphere. Furthermore, the Southern Hemisphere has more lands toward equator compared to the very Southern side of the Earth. With this knowledge, I would like to discuss about the weather trends including humidity, wind speed, and temperature depending on latitudes. 

In the ‘Latitude vs. Temperature Plot’, it represents the relationship between latitude and max temperature across all latitudes. Not surprisingly, the places near equator tend to be hottest around the Earth. It is also found the tendency that the temperature has plummeted as the latitude increases in Northern Hemisphere. Otherwise, the Southern Hemisphere does not show the drastic decrease in terms of temperature as the places go further away from the equator. In other words, it shows quite mild change in temperature when it comes to the places in the Southern Hemisphere. 

As I mentioned earlier, it would happen because the Southern Hemisphere’s surface area has more water. By having more ocean, water tends to keep the heat longer than land. So, it helps the Southern Hemisphere to be warmer in general and its temperature gap depending on the latitude is not as huge as the Northern Hemisphere. This tendency is also appeared in the linear regression as well. Both Northern and Southern Hemisphere have shown the quite strong correlation between its latitude and the temperature. However, Northern Hemisphere has stronger linear relation.  

In terms of humidity, the areas around equator, specifically between -20 and +20 degree of latitude, tend to be drier than other areas in general. This would explain why lots of desserts can be found in those areas. Even though it is not a huge relation, we can also find that a place on the higher latitude is likely to be more humid in the Northern Hemisphere. In the Southern Hemisphere, it also shows slight increase of humidity as it is closer to equator. However, the correlation value is 0.25 while the r-squared value is 0.05. Accordingly, the relationship between the humidity and its latitude in the Southern Hemisphere cannot be significant enough to draw a meaningful conclusion. Instead, there would potentially be other factors such as humidity of the air at saturation depending on the temperature and water content that influences on humidity depending on the region. 

In the data, it contains how cloudiness could be related to its latitude and there are weak but positive correlations found in both Hemispheres. Up to about 40 degree of Latitude, there seem to be more cloud-free areas while there are more clouds in the areas above 40 degree of Latitude. The r-squared value could be low in this case since there are quite a significant disparity from 0 to 100. Likewise, the Southern Hemisphere shows similar trends. Up to about -20 degree of Latitude, there are more places recorded as no cloud while its cloudiness of many regions above -20 to equator show 100% of clouds. Due to this disparity, its r-squared value is also low.

The relationship between wind speed and latitude is also shown in the analysis. Their linear regression and correlation coefficient are not significant to link how they rely on one another. I believe there are other variables including Earth’s rotation and its geographical characteristics that influence on the wind.  

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import scipy.stats as st
import json

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy
from pprint import pprint

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
print(weather_api_key)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Create an empty list to store the data 
city_list = [] 
city_latitudes = []
city_longitudes = []
city_clouds = []
city_humidity = []
city_max_temp = []
city_wind_speed = []
city_country = []
city_date = []
index_counter = 0 
set_counter = 1 
```

```python
print("Beginning Data Retrieval ")
print("-----------------------------")

base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{base_url}appid={weather_api_key}&units={units}&q="


# For each city name in cities list, do below things...
for index, city in enumerate(cities, start = 1):
    try:
        response = requests.get(query_url + city).json()
        city_list.append(response["name"])
        city_clouds.append(response["clouds"]["all"])
        city_country.append(response["sys"]["country"])
        city_date.append(response["dt"])
        city_humidity.append(response["main"]["humidity"])
        city_latitudes.append(response["coord"]["lat"])
        city_longitudes.append(response["coord"]["lon"])
        city_max_temp.append(response['main']['temp_max'])
        city_wind_speed.append(response["wind"]["speed"])
        if index_counter > 49:
            index_counter = 0
            set_counter = set_counter + 1
    
        else:
            index_counter = index_counter + 1
            
        print(f"Processing Record {index_counter} of Set {set_counter} : {city}") 
  
    except(KeyError, IndexError):
        print("City not found. Skipping...")

print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
city_df = pd.DataFrame({
    'City Name': city_list,
    'Latitude':city_latitudes,
    'Longitude': city_longitudes,
    'Max Temp': city_max_temp,
    'Humidity' :city_humidity,
    'Cloudiness': city_clouds,
    'Wind Speed': city_wind_speed,
    'Country' : city_country,
    "Date": city_date 
})
city_df
```

```python
city_df.to_csv('citydata.csv', index= False)
```

## Linear Regression

```python
#Create new DataFrames to divide northern and sourthern hemisphere
northern_hemisphere = clean_city_data.loc[pd.to_numeric(clean_city_data["Latitude"]).astype(float) > 0, :]
southern_hemisphere = clean_city_data.loc[pd.to_numeric(clean_city_data["Latitude"]).astype(float) < 0, :]

# northern_hemisphere = clean_city_data.loc[clean_city_data["Latitude"] >= 0]
# southern_hemisphere = clean_city_data.loc[clean_city_data["Latitude"] < 0]
# print(northern_hemisphere)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
fig,ax = plt.subplots(figsize=(12,6))
x_values = pd.to_numeric(northern_hemisphere['Latitude']).astype(float)
y_values = pd.to_numeric(northern_hemisphere['Max Temp']).astype(float)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values,y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " +str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=12,color="red")
plt.xlabel("Latitude")
plt.ylabel("Max Temp")
plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")

#print the statistical values to see the relationship between the two variables
correlation = st.pearsonr(x_values,y_values)
print(f"The correlation coefficient is {round(correlation[0],2)}")
print(f"Regression line equation is: {line_eq}")
print(f"The r-squared is: {rvalue**2}")


plt.show()
```
